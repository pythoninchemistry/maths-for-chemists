# Integrating Trigonometric Functions

We can recall that: 

$$
\textrm{If }y=\sin(ax)\textrm{ then }\frac{dy}{dx}=a\cos(ax)
$$
$$
\textrm{If }y=\cos(ax)\textrm{ then }\frac{dy}{dx}=-a\sin(ax)
$$

When integrating, we do the opposite of differentiation, so: 

$$
\int\cos(ax)\;dx = \frac{\sin(ax)}{a} + C
$$
$$
\int\sin(ax)\;dx = \frac{-\cos(ax)}{a} + C
$$

```{admonition} Example
:class: example

Find the following: 

1. $\int\cos(4x)\;dx$
2. $\int 6\sin(3x)\;dx$
3. $\int\left(5\cos(-x) + \sin(3x)\right)\;dx$

*Solution:* 

1. Using the rule $\int\cos(ax)\;dx = \frac{\sin(ax)}{a} + C$, we can find: 

    $$
    \int\cos(4x)\;dx = \frac{\sin(4x)}{4} + C
    $$

2. First, we should note $\int 6\sin(3x)\;dx= 6\int \sin(3x)\;dx$. 
    Now we can apply the rules above to get: 

    $$
    6\int \sin(3x)\;dx = 6\times \frac{-\cos(3x)}{3} + C = -2\cos(3x)
    $$

3. Splitting up the integral and extracting constants gives: 

    $$
    \begin{aligned}
    \int \left(5\cos(-x) + \sin(3x)\right)\;dx & = 5\int \cos(-x)\;dx + \int \sin(3x)\;dx \\
    & 5\times \frac{\sin(-x)}{-1} + \frac{-\cos(3x)}{3} + C \\
    & -5 \sin(-x) - \frac{\cos(3x)}{3} + C
    \end{aligned}
    $$
```

As one would, hopefully, expect, `sympy` can handle these operations. 

`````python
from sympy import symbols, sin, cos, integrate

x = symbols('x')

integrate(cos(4 * x))
`````

`````python
integrate(6 * sin(3 * x))
`````
